```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("house_data.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df["area"].fillna(df["area"].mode()[0],inplace=True)
df["halls"].fillna(df["halls"].mode()[0],inplace=True)
df["bedrooms"].fillna(df["bedrooms"].mode()[0],inplace=True)
df["age"].fillna(df["age"].mode()[0],inplace=True)
df["attached_washroom"].fillna(df["attached_washroom"].mode()[0],inplace=True)
df["floors"].fillna(df["floors"].mode()[0],inplace=True)
df["price"].fillna(df["price"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
df.head(5)
```

```python
model=LinearRegression()
x=df[["area","halls","bedrooms","age","attached_washroom","floors"]]
y=df["price"]
model.fit(x,y)
```

```python
#                          area	 halls	bedrooms   age	attached_washroom	floors
prediction=model.predict([[8977,  3,      3,        18,         2,             5]])
print(" if area=8977, halls=3, bedrooms= 3, age	= 18, attached_washroom=2, floors=5 then price =",prediction)
```
